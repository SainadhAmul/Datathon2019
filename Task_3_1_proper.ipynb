```python
import pandas as pd
import numpy as np
import os
import re
```

```python
os.chdir('C://Users//Administrator//Desktop//datasets-v5//tasks-2-3//train')


filenames = os.listdir()
#print(filenames)
```

```python
txt_files = list(filter(lambda x: x[-4:] == '.txt', filenames))
lab_files = list(filter(lambda x: x[-12:] == 'task3.labels', filenames))
```

```python
task3_train=pd.DataFrame()
task3_train.head()
```

```python
for i in range(len(lab_files)):
    lab_list =open(lab_files[i]).readlines()
    txt_list = open(txt_files[i],encoding='utf-8').readlines()
    for lab in lab_list:
        l1 =[]
        #print(lab)
        index_one = int((lab.split('\t')[2]))
        index_two = int((lab.split('\t')[3]))
        print(index_one,index_two)
        #print(txt_files[i])
        #txt =str(txt_list).replace(' ', "")
        mm=[]
        for tt in txt_list:
            #tt1 =" ".join(re.findall("[a-zA-Z]+", tt))
            mm=mm+list(tt)
        #print(mm)
        txt = mm[index_one:index_two]
        txtt = "".join(str(x) for x in txt).replace('\n','').replace('"','')
        curr_row={
            'article_id':lab.split('\t')[0],
            'news_text':txtt,
            'label':lab.split('\t')[1]
            
        }
        task3_train = task3_train.append(curr_row,ignore_index=True)    
    #lable_df = pd.concat([lable_df,df],axis =0 ,ignore_index = True)
  
   
        
           
task3_train.head(20)    

    
        
```
